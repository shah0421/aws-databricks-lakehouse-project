# Monthly Order Summary
For each of the customer, produce the following summary per month

1. total orders
2. total items bought
3. total amount spend

```python
%sql
CREATE OR REPLACE TABLE gizmobox.gold.order_monthly_summary
SELECT 
  o.customer_id,
  date_format(o.transaction_timestamp, 'yyyy-MM') as order_month,
  count(distinct o.order_id) as total_orders,
  sum(o.quantity) as total_items,
  sum((o.price*o.quantity) - COALESCE(r.refund_amount,0)) as total_paid
FROM gizmobox.silver.orders o
JOIN gizmobox.silver.payments p on o.order_id = p.order_id
LEFT JOIN hive_metastore.silver.refunds r on p.payment_id = r.payment_id
GROUP BY 
  o.customer_id,
  order_month
order by customer_id;
```

```python
%sql
SELECT * FROM gizmobox.gold.order_monthly_summary;
```
