## 2. Create External Table

```python
%sql
DROP TABLE IF EXISTS gizmobox.bronze.payments;
CREATE TABLE gizmobox.bronze.payments
USING CSV
Options (
  header="true",
  delimiter=","
)
LOCATION 's3a://amzn-s3-gizmobox-project-bucket/gizmobox/landing/external-data/Payments/'
```

```python
%sql
SELECT * FROM gizmobox.bronze.payments;
```

```python
%sql
DESCRIBE EXTENDED gizmobox.bronze.payments;
```

## 3. Demonstrate the effect of Adding/Updating/Deleting files

```python
%sql
REFRESH TABLE gizmobox.bronze.payments;
-- any update in the source will cause error when trying to access the external table. executing the Refresh command fixes the metadata of the source and let's us access the data again.  
```

## 4. Demonstrate the effect of Dropping the Table

```python
%sql
DROP TABLE IF EXISTS gizmobox.bronze.payments;
-- S3 file data for this table will not be lost. only the metadata of this external table in databricks will be lost.
```
