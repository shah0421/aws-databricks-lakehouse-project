# Transform Memberships Data
1. Extract customer_id from the file path
2. Write transformed data to the Silver schema

## 1. Extract customer_id from the file path

```python
%sql
SELECT * FROM gizmobox.bronze.v_memberships limit 2;
```

```python
%sql
SELECT 
  regexp_extract(path, '/.*/([0-9]+).\\.png$', 1) AS custome_id,
  content as membership_card
FROM gizmobox.bronze.v_memberships;
```

## 2. Write transformed data to the Silver schema

```python
%sql
CREATE OR REPLACE TABLE gizmobox.silver.memberships
AS
SELECT 
  regexp_extract(path, '/.*/([0-9]+).\\.png$', 1) AS custome_id,
  content as membership_card
FROM gizmobox.bronze.v_memberships;
```
